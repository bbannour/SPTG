# Thermostat Example

```python
from pyversity import *
```

```python
system = System("ThermostatSystem", timed="discrete")
```

Create the state machines contained in the system

```python
thermostat = system.addStateMachine("Thermostat")
equipment = system.addStateMachine("Equipment")
```

State variables and ports for Thermostat machine

```python
t = thermostat.addIntVar("t")
y = thermostat.addIntVar("y")
z = thermostat.addIntVar("z")

TMIN = thermostat.addIntConst("TMIN", 0)
TMAX = thermostat.addIntConst("TMAX", 25)

HEATING = thermostat.addIntConst("HEATING", 1)
COOLING = thermostat.addIntConst("COOLING", 2)
OFF = thermostat.addIntConst("OFF", 3)

thermostat_error = thermostat.addInPort('err', params=['int'])
thermostat_equip = thermostat.addInPort('equip', params=['int'])
sens = thermostat.addInPort('sens', params=['int'])

screen = thermostat.addOutPort('screen', params=['int'])
thermostat_dt = thermostat.addOutPort('dt', params=['int'])
test = thermostat.addOutPort('test', params=['int'])

print(thermostat_error)
print(system)
```

Add states to the state machines

```python
qinit = thermostat.addInitialState("qinit")
q0 = thermostat.addState("q0")
q1 = thermostat.addState("q1")
q2 = thermostat.addState("q2")
q3 = thermostat.addState("q3")
q4 = thermostat.addState("q4")
q5 = thermostat.addState("q5")
q6 = thermostat.addState("q6")
q7 = thermostat.addState("q7")

sinit = equipment.addInitialState("sinit")
s0 = equipment.addState("s0")
s1 = equipment.addState("s1")
s2 = equipment.addState("s2")
s3 = equipment.addState("s3")
s4 = equipment.addState("s4")
```

Add transition to the Thermostat state machine

```python
tinit = qinit.addTransition("tinit", q0)

t1 = q0.addTransition("t1", q0)
t1.addReceive(thermostat_error, ["z"])

t2 = q0.addTransition("t2", q0)
t2.addReceive(thermostat_equip, ["y"])

t3 = q0.addTransition("t3", q1)
t3.addReceive(sens, ["t"])

t11 = q1.addTransition("t11", q2)
t11.addSend(screen, ["HEATING"])
t11.addGuard("t < TMIN")

t12 = q1.addTransition("t12", q5)
t12.addSend(screen, ["COOLING"])
t12.addGuard("t > TMAX")

t13 = q1.addTransition("t13", q0)
t13.addSend(screen, ["OFF"])
t13.addGuard("t >= TMIN && t <= TMAX")

t21 = q2.addTransition("t21", q2)
t21.addReceive(thermostat_error, ["z"])

t22 = q2.addTransition("t22", q3)
t22.addSend(thermostat_dt, ["TMIN-t"])

t31 = q3.addTransition("t31", q3)
t31.addReceive(thermostat_equip, ["y"])

t32 = q3.addTransition("t32", q3)
t32.addReceive(thermostat_error, ["z"])

t33 = q3.addTransition("t33", q4)
t33.addReceive(sens, ["t"])

t41 = q4.addTransition("t41", q4)
t41.addReceive(thermostat_equip, ["y"])

t42 = q4.addTransition("t42", q4)
t42.addReceive(thermostat_error, ["z"])

t43 = q4.addTransition("t43", q0)
t43.addSend(thermostat_dt, ["0"])
t43.addGuard("t >= TMIN")

t44 = q4.addTransition("t44", q2)
t44.addGuard("t < TMIN")

t51 = q5.addTransition("t51", q5)
t51.addReceive(thermostat_error, ["z"])

t52 = q5.addTransition("t52", q6)
t52.addSend(thermostat_dt, ["TMAX-t"])

t61 = q6.addTransition("t61", q6)
t61.addReceive(thermostat_equip, ["y"])

t62 = q6.addTransition("t62", q6)
t62.addReceive(thermostat_error, ["z"])

t63 = q6.addTransition("t63", q7)
t63.addReceive(sens, ["t"])

t71 = q7.addTransition("t71", q7)
t71.addReceive(thermostat_equip, ["y"])

t72 = q7.addTransition("t72", q7)
t72.addReceive(thermostat_error, ["z"])

t73 = q7.addTransition("t73", q0)
t73.addSend(thermostat_dt, ["0"])
t73.addGuard("t <= TMAX")

t74 = q7.addTransition("t74", q5)
t74.addGuard("t > TMAX")
```

State variables and ports for Thermostat machine

```python
x = equipment.addIntVar("x")
isOn = equipment.addBoolVar("isOn")
c = equipment.addClock("c")

equipment_dt = equipment.addInPort('dt', params=['int']) 
equipment_equip = equipment.addOutPort('equip', params=['int'])
equipment_error = equipment.addOutPort('err', params=['int'])

TIMEOUT = equipment.addIntConst("TIMEOUT", 0)
```

Add transition to the Equipment state machine

```python
einit = sinit.addTransition("sinit", s0)

e1 = s0.addTransition("e1", s1)
e1.addReceive(equipment_dt, ["x"])
e1.addStatement("c = 0")

e11 = s1.addTransition("e11", s0)
e11.addGuard("x==0")
e11.addClockGuard("c <= 2")
e11.addStatement("c = 0")
e11.addSend(equipment_equip, ["0"])
e11.addStatement("isOn = false")

e12 = s1.addTransition("e12", s2)
e12.addGuard("x>0")
e12.addClockGuard("c <= 2")
e12.addStatement("c = 0")
e12.addStatement("isOn = true")

e13 = s1.addTransition("e13", s4)
e13.addClockGuard("c > 2")
e13.addSend(equipment_error, ["TIMEOUT"])

e14 = s1.addTransition("e14", s2)
e14.addGuard("x<0")
e14.addClockGuard("c <= 2")
e14.addStatement("c = 0")
e14.addStatement("isOn = true")

e21 = s2.addTransition("e21", s0)

e22 = s2.addTransition("e22", s2)
e22.addGuard("x>0")
e22.addClockGuard("c <= 1")
e22.addStatement("c = 0")
e22.addSend(equipment_equip, ["4"])
e22.addStatement("x = x-4")

e23 = s2.addTransition("e23", s4)
e23.addClockGuard("c > 1")
e23.addSend(equipment_error, ["TIMEOUT"])

e31 = s3.addTransition("e31", s0)

e32 = s3.addTransition("e32", s3)
e32.addGuard("x<0")
e32.addClockGuard("c <= 1")
e32.addStatement("c = 0")
e32.addSend(equipment_equip, ["-4"])
e32.addStatement("x = x+4")

e33 = s3.addTransition("e33", s4)
e33.addClockGuard("c > 1")
e33.addSend(equipment_error, ["TIMEOUT"])

e41 = s4.addTransition("e41", s1)
e41.addReceive(equipment_dt, ["x"])


display(system)
```

Connect ports together

```python
system.connectEnv(sens)
system.connectEnv(screen)
system.connectRdv(thermostat_dt, equipment_dt)
system.connectRdv(thermostat_error, equipment_error)
system.connectRdv(thermostat_equip, equipment_equip)

print(system)
```

## Create your configuration

Configuration is holding the general parameters of the computation.

```python
c = Configuration()
c.setWorkspace(rootdir='./', outdir='./')
c.setStrategy('BFS')
c.setSolver('Z3')
c.setVerbosity('SILENT')
c.setDevOpts(level='HIGH', log='monLog.log', debug='myDebug.dbg')
#c.setDevFlags('GOD_MODE')
c.check()
print(c)
```

## Engine : la classe mère

```python
engine = Engine(c)
print(engine)
ok = engine.setSystem(system)
print ("Set system:", ok)
result = engine.start()


print(result)
```
